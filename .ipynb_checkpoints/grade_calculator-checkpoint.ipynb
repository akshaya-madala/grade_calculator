#  Student Grade Calculator

## Project Objective
Build a Python program that:
- Accepts student name and marks.
- Calculates grade using `if-elif-else`.
- Validates marks (0–100).
- Uses a function.
- Displays an encouraging message based on the grade.

```python
print("STUDENT GRADE CALCULATOR")
print("Grading Rules:")
print("A : 90 - 100")
print("B : 80 - 89")
print("C : 70 - 79")
print("D : 60 - 69")
print("F : 0 - 59")
```

# Get Student Details

This section accepts:
Student Name,
Student Marks

```python
student_name = input("Enter Student Name: ")
marks = int(input("Enter Marks: "))

print("Student Name:", student_name)
print("Marks:", marks)
```

# Grading Logic

The program uses if-elif-else statements to determine the student's grade.

```python
if marks >= 90:
    grade = "A"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```

#  Input Validation

The program uses a while loop to ensure the marks entered are between 0 and 100.

```python
while True:
    marks = int(input("Enter Marks (0-100): "))

    if 0 <= marks <= 100:
        break
    else:
        print("Invalid Marks! Please enter marks between 0 and 100.")

print("Valid Marks:", marks)
```

# Function and Final Result

A function is created to:
- Calculate the grade.
- Return the grade and an encouraging message.

```python
# Function to calculate grade
def calculate_grade(marks):
    if marks >= 90:
        return "A", "Excellent! Outstanding Performance! "
    elif marks >= 80:
        return "B", "Very Good! Keep it up! 👍"
    elif marks >= 70:
        return "C", "Good Job! Keep improving! 😊"
    elif marks >= 60:
        return "D", "You Passed! Work harder! 💪"
    else:
        return "F", "Don't give up! Practice more! "


# Get student name
student_name = input("Enter student name: ")

# Input validation
while True:
    marks = int(input("Enter marks (0-100): "))

    if 0 <= marks <= 100:
        break
    else:
        print("Invalid input! Please enter marks between 0 and 100.")

# Calculate grade
grade, message = calculate_grade(marks)

# Final Output
print()
print(f"📊RESULT FOR {student_name.upper()}:")
print(f"Marks: {marks}/100")
print(f"Grade: {grade}")
print(f"Message: {message}")
```
